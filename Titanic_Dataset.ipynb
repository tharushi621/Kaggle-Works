```python
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv('tested.csv')
df.head()
```

```python
df.drop(['Name','PassengerId','SibSp','Parch','Ticket','Cabin','Embarked','Fare'],axis='columns',inplace=True )
df.head()
```

```python
iv=df.drop('Survived',axis='columns')
dv=df.Survived
```

```python
iv
```

```python
dv
```

```python
dummies=pd.get_dummies(iv.Sex,dtype=int)
```

```python
dummies.head(200)
```

```python
iv=pd.concat([iv,dummies],axis='columns')
```

```python
iv.head(3)
```

```python
iv.drop(['Sex','male'],axis='columns',inplace=True)
```

```python
iv.head(3)
```

```python
iv.columns[iv.isna().any()]
```

```python
median_age=math.floor(df.Age.median())
print(median_age)
```

```python
iv['Age'].fillna(value=median_age,inplace=True)
```

```python
iv.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(iv,dv,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
